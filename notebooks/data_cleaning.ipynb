## Genre Transformation

In this step, the original `genres` column from the movies dataset is transformed into a normalized structure suitable for relational analysis.

### Process

- Selected only relevant columns (`id`, `genres`)
- Removed rows with missing genre values
- Renamed `id` to `movie_id` for consistency
- Split multiple genres into lists using comma separation
- Expanded the dataset so that each row contains a single genre per movie
- Cleaned genre names by removing extra spaces
- Renamed the column to `genre_name`
- Removed duplicate records
- Reset the index for a clean structure
- Converted `movie_id` to integer type

### Result

The resulting dataset contains a normalized many-to-one relationship between movies and genres, making it suitable for SQL joins and analytical queries.

### Output

The cleaned dataset is exported as:

`genre_film_clear.csv`

```python
movie = pd.read_csv("File puliti\movies_clean.csv", encoding="utf-8")
genre_film = movie[["id", "genres"]].copy()
genre_film = genre_film.dropna(subset=["genres"])
genre_film = genre_film.rename(columns={"id": "movie_id"})
genre_film["genres"] = genre_film["genres"].str.split(",")
genre_film = genre_film.explode("genres")
genre_film["genres"] = genre_film["genres"].str.strip()
genre_film = genre_film.rename(columns={"genres": "genre_name"})
genre_film = genre_film.drop_duplicates()
genre_film = genre_film.reset_index(drop=True)
genre_film['movie_id']=genre_film['movie_id'].astype(int)
genre_film.to_csv("genre_film_clear.csv", index=False, encoding="utf-8")
```

## Data Export

In this final step, the cleaned and transformed datasets are exported as CSV files.

These files will be used for:
- SQL database integration
- Further data analysis
- Business Intelligence tools (e.g. Power BI, Looker)

This step completes the ETL process, making the data ready for downstream applications.

```python
import pandas as pd
from sqlalchemy import create_engine

#connessione MySQL
engine = create_engine("mysql+pymysql://root:211100@localhost/Movies_Database")

# carica csv
cast = pd.read_csv("File puliti\cast_clean.csv")
crew = pd.read_csv("File puliti\crew_clean.csv")
genres=pd.read_csv("File puliti\genres_clean.csv")
movies= pd.read_csv("File puliti\movies_clean.csv")
reviews= pd.read_csv("File puliti/reviews_clean.csv")

#invia a MySQL
cast.to_sql("cast_clean", con=engine, if_exists="replace", index=False)
crew.to_sql("crew_clean", con=engine, if_exists="replace", index=False)
genres.to_sql("genres_clean", con=engine, if_exists="replace", index=False)
movies.to_sql("movies_clean", con=engine, if_exists="replace", index=False)
reviews.to_sql("reviews_clean", con=engine, if_exists="replace", index=False)
genre_film.to_sql("genre_film_clean", con=engine, if_exists="replace", index=False)

```
